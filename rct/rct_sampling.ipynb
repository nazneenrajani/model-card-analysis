## Select models with high downloads in 2021 or 2022 with no model card and model card with <300 words.

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/nazneenrajani/workspace/model-card-analysis/model_top5k_sep30.csv', usecols=range(1,12)) 

df_nomc = df.loc[df['has_modelcard'] == "No"]
df_smc = df[df.iloc[:,6] == "Yes"]

df_mc = pd.concat([df_nomc, df_smc], ignore_index=True)

df_top = df_mc.sort_values(by=['downloads'], ascending=False, ignore_index=True)[:1000] # sort by downloads and take top N models that satisfy our criteria of having ~300 models after fitlering.
```

```python
df_notest = df_top.drop(df_top[df_top['model_name'].str.contains('test')].index)
df_top = df_notest
```

```python
df_topsample = df_top[df_top['creation_time'].str.contains('2022', na=False) | df_top['creation_time'].str.contains('2021', na=False)]
```

```python
df_top_no_modelcard = df_topsample[df_topsample.iloc[:,3] == "No"]
df_top_short_modelcard = df_topsample[df_topsample.iloc[:,6] == "Yes"][:80]
```

```python
df_no_modelcard_treatment = df_top_no_modelcard.sample(n=60) #sample 60 models without model cards
df_no_modelcard_control = df_top_no_modelcard.drop(df_no_modelcard_treatment.index)
```

```python
df_short_modelcard_treatment = df_top_short_modelcard.sample(n=40) #sample 40 models with model cards that are <=100 words
df_short_modelcard_control = df_top_short_modelcard.drop(df_short_modelcard_treatment.index)
```

## Batch the treatment group into 2

```python
df_no_modelcard_treatment_batch1 = df_no_modelcard_treatment.sample(n=30)
df_no_modelcard_treatment_batch2 = df_no_modelcard_treatment.drop(df_no_modelcard_treatment_batch1.index)

df_short_modelcard_treatment_batch1 = df_short_modelcard_treatment.sample(n=20)
df_short_modelcard_treatment_batch2 = df_short_modelcard_treatment.drop(df_short_modelcard_treatment_batch1.index)
```

```python
df_short_modelcard_treatment_batch1.sort_values(by=['model_name'])
```

```python
df_no_modelcard_treatment_batch1.to_csv('treatment_no_modelcard_batch1.csv', index=False)
df_short_modelcard_treatment_batch1.to_csv('treatment_short_modelcard_batch1.csv', index=False)

df_no_modelcard_treatment_batch2.to_csv('treatment_no_modelcard_batch2.csv', index=False)
df_short_modelcard_treatment_batch2.to_csv('treatment_short_modelcard_batch2.csv', index=False)

df_no_modelcard_control.to_csv('control_no_modelcard.csv', index=False)
df_short_modelcard_control.to_csv('control_short_modelcard.csv', index=False)
```
